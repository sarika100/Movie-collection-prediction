```python
len(columns)
```

```python
def outlier_treatment(data, columns):
    for col in columns:
        q1=data[col].quantile(0.25)
        q3=data[col].quantile(0.75)
        iqr=q3-q1
        lf=q1-1.5*iqr
        uf=q3+1.5*iqr
        outlier_condition=(data[col]<lf)|(data[col]>uf)
        data.loc[outlier_condition, col]=data[col].median()
    return data
    
```

```python
df= outlier_treatment(data,columns)
```

```python
from tensorflow.keras.models import Sequential  # enables us to add layers into our neural network 
from tensorflow.keras.layers import Dense       # then we add dense layer to ensure all neural networks are connected to each other
```

```python
from sklearn.preprocessing import StandardScaler # to reduce complexity, makes data smaller numerically
sc = StandardScaler()
```

```python
scaled_x = sc.fit_transform(X_train) #converting the data on scale of -1 to +1
```

```python
scaled_x
```

```python
input_neuron = scaled_x.shape[1]  # shape is feature, 1 means column
```

```python
hidden_neuron = 32  # a little bit more than input neuron (more than 20)
output_neuron = 1   # regression so we need only 1 value in the output
```

```python
model = Sequential()
```

```python
model.add(Dense(hidden_neuron, activation = 'relu', input_shape=(input_neuron,))) #input layer
```

```python
model.add(Dense(hidden_neuron, activation = 'relu')) #hidden layer
model.add(Dense(hidden_neuron//2, activation = 'relu')) #hidden layer
model.add(Dense(hidden_neuron//2, activation = 'relu')) #hidden layer
```

```python
model.add(Dense(output_neuron, activation = 'linear')) #hidden layer
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['mse']) # adam = adaptive gradient descent with momentum
# momentum is fractional component of previous slope, reduces effort, time and complexity. 
```

```python
model.fit(scaled_x, Y_train, epochs = 150, batch_size = 32, validation_split = 0.2)
```
